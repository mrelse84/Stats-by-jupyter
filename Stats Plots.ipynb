# Stats Plots

## 1. 표준 정규 분포와 t분포 그리기

```python
rv_norm = stats.norm(0,1)
rv_t15 = stats.t(15-1)
rv_t7 = stats.t(7-1)
x = np.linspace(-5, 5, 100)
pdf_norm = rv_norm.pdf(x)
pdf_t15 = rv_t15.pdf(x)
pdf_t7 = rv_t7.pdf(x)
plt.figure(figsize=(12, 5))
plt.plot(x, pdf_norm, 'k-',lw=1, label='Norm(0,1)')
plt.plot(x, pdf_t15, 'b-',lw=1, label='t(df=14)')
plt.plot(x, pdf_t7, 'r-',lw=1, label='t(df=6)')
plt.title('Norm and t dist')
plt.xlabel('x')
plt.ylabel('pdf')
plt.grid(True)
plt.legend()
```

### 1.1 Calc ppf & cdf

ppf : percent point function (inverse of cdf)  
cdf : cumulative density function

```python
# 신뢰 수준 95% (유의 수준 0.05)
alpha = 0.05
rv_norm.ppf(1-alpha/2) # 양측 임계값을 구하기 위해 유의 수준을 2로 나눈다
```

```python
# 양측 임계값 1.96를 얻었다. 이 것의 좌측 임계값은 -1.96이고 우측 임계값은 +1.96이다.
```

```python
# 양측 임계값 1.96의 (유의) 확률 (p-value)
2*(1-rv_norm.cdf(1.96)) # 우측 임계값 +1.96의 (유의) 확률 (p-value)를 구한 후 2를 곱한다
```

## 2. F분포 그리기

```python
rv_f = stats.f(4,50)
rv_f_r = stats.f(50,4)
rv_f_1_10 = stats.f(1,10)
rv_f_10_1 = stats.f(10,1)
x = np.linspace(0, 5, 100)
pdf_f = rv_f.pdf(x)
pdf_f_r = rv_f_r.pdf(x)
pdf_f_1_10 = rv_f_1_10.pdf(x)
pdf_f_10_1 = rv_f_10_1.pdf(x)
plt.figure(figsize=(12, 5))
plt.plot(x, pdf_f, 'b-',lw=1, label='F(4,50)')
plt.plot(x, pdf_f_r, 'r-',lw=1, label='F(50,4)')
plt.plot(x, pdf_f_1_10, 'b--',lw=1, label='F(1,10)')
plt.plot(x, pdf_f_10_1, 'r--',lw=1, label='F(10,1)')
plt.title('F dist')
plt.xlabel('F')
plt.ylabel('pdf')
plt.grid()
plt.legend()
```

## 3. $\chi^2$분포 그리기

```python
rv_c2_1 = stats.chi2(1)
rv_c2_5 = stats.chi2(5)
rv_c2_10 = stats.chi2(10)
x = np.linspace(0, 50, 100)
pdf_c2_1 = rv_c2_1.pdf(x)
pdf_c2_5 = rv_c2_5.pdf(x)
pdf_c2_10 = rv_c2_10.pdf(x)
plt.figure(figsize=(12, 5))
plt.plot(x, pdf_c2_1, 'k-',lw=1, label='$\chi^2$(1)')
plt.plot(x, pdf_c2_5, 'b-',lw=1, label='$\chi^2$(5)')
plt.plot(x, pdf_c2_10, 'r-',lw=1, label='$\chi^2$(10)')
plt.title('$\chi^2$ dist')
plt.xlabel('$\chi^2$')
plt.ylabel('pdf')
plt.grid()
plt.legend()
```
